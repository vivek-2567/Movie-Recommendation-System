```python
import pandas as pd
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
movies = movies.merge(credits,on='title')
movies
```

```python
def cnvt(given):
    ans = []
    x = eval(given)
    for i in x:
        ans.append(i['name'])
    return ans
```

```python
movies['genres'] = movies['genres'].apply(cnvt)
movies
```

```python
movies = movies[['title','genres','id','vote_count','vote_average']]
movies
```

```python
checking = movies
ans = pd.DataFrame(columns=['title','genres','id','vote_count','vote_average'])
selection = ['Action','Fantasy','Romance']
for i in selection:
    for j in range(len(checking)):
        try:
            if i in checking['genres'][j]:
                ans = pd.concat([ans,checking.loc[j:j]])
        except:
            pass
    checking = ans
ans.head(10)
```

```python
checking = movies
checking = checking.drop('genres',1).join(checking.genres.str.join('|').str.get_dummies())
checking
```

```python
checking.to_csv("Filtering_movies.csv")
```

```python
def best_one(df):
    import math
    x1 = math.exp(df['vote_count']//1000)
    x2 = df['vote_count']%1000 * 7
    x3 = df['vote_average']*13
    return math.log(x1+x2+x3)
```

```python
def select(selection):
    df = pd.read_csv('Filtering_movies.csv')
    for i in selection:
        df = df[df[i]==1]

    df = df[['id','title','vote_count','vote_average']]
    df['sorting_column'] = df.apply(lambda x: best_one(x),axis = 1)
    return df.sort_values('sorting_column',ascending=False).head(10)
```

```python
select(['Action','Fantasy','Romance'])
```
