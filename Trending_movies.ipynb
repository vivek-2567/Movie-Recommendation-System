```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
movies.head()
```

```python
y = movies[movies['vote_average']>=7.7]
top_50_movies = y[y['vote_count']>3500].sort_values('vote_average',ascending=False).head(50)
top_50_movies[['title','id']]
```

```python

top_50_movies.to_csv("Top_50_movies.csv")
```
